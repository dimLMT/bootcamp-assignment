2. What query would you run to get all the countries that speak Greek? Your query should return the name of the country, language, and language percentage.
```python
q = """
SELECT continent, COUNT(*) AS `number of countries` FROM countries
GROUP BY continent
ORDER BY COUNT(*) DESC;
"""
pd.read_sql(text(q), conn)
```

```python
q = """
SELECT name, language, percentage FROM countries
JOIN languages ON languages.country_id=countries.id
WHERE language='Greek'
ORDER BY percentage DESC;
"""
pd.read_sql(text(q), conn)
```

- What query would you run to get all the cities of Mexico inside the Veracruz district and have a population greater than 100,000? The query should return the Country Name, City Name, District, and Population. - ​6. What query would you run to get all the ten cities in Brazil with the smallest population? Your query should include the country name, city name, and population. Arrange the result by population in ascending order. Limit the result to the 10 cities with the smallest populat
- W What query would you run to get all languages spoken greater than 10% in Monaco? Display the country name, language, and percntage.​


```python
q = """
SELECT countries.name AS 'Country Name', cities.name AS City, cities.population, cities.district
FROM countries
JOIN cities ON countries.id = cities.country_id
WHERE countries.name = 'Mexico' AND cities.population > 100000 AND cities.district='Veracruz' 
ORDER BY cities.population DESC;
"""
# Pass the query though the text function before running read_sql
pd.read_sql(text(q), conn)
```

```python
q = """
SELECT countries.name AS 'Country Name', cities.name AS City, cities.population
FROM countries
JOIN cities ON countries.id = cities.country_id
WHERE countries.name = 'Brazil '
ORDER BY cities.population 
LIMIT 10;
"""
# Pass the query though the text function before running read_sql
pd.read_sql(text(q), conn)
```

```python
q = """
SELECT countries.name AS 'Country Name', language, percentage FROM countries
JOIN languages ON languages.country_id=countries.id
WHERE countries.name = 'Monaco' AND percentage>10;
"""
# Pass the query though the text function before running read_sql
pd.read_sql(text(q), conn)
```
